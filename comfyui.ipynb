#🚀 Abrir o Túnel e Ligar ComfyUI

```python
# Commented out IPython magic to ensure Python compatibility.
# 1. Entra na pasta e limpa o que estiver travado
# %cd /content/ComfyUI
!pkill -f main.py
!pkill cloudflared

# 2. Roda o ComfyUI em SEGUNDO PLANO (Background)
# O "> /dev/null 2>&1 &" serve para não poluir sua tela e liberar a célula
print("🚀 Ligando o ComfyUI no 'escuro'...")
!python main.py --dont-print-server > /dev/null 2>&1 &

# 3. Espera o servidor "acordar"
import time
time.sleep(15)

# 4. Roda o Cloudflare na mesma célula
print("🔗 Gerando o link do túnel...")
!wget -q -nc https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64.deb
!dpkg -i cloudflared-linux-amd64.deb > /dev/null 2>&1
!cloudflared tunnel --url http://127.0.0.1:8188
```

```python
# @title 🚀 LANÇAMENTO DUPLO (Motor + Túnel)
import os, time

# 1. Limpa tudo que tá travado pra começar do zero
!pkill -f main.py
!pkill cloudflared

# 2. Liga o Túnel em segundo plano (o & no final faz ele rodar "escondido")
print("🔗 Iniciando Túnel...")
!nohup cloudflared tunnel --url http://127.0.0.1:8188 > tunnel.log 2>&1 &

# 3. Espera 5 segundos e pega o link
time.sleep(5)
print("\n--- SEU LINK NOVO (ABRA NUMA ABA ANÔNIMA) ---")
!grep -o 'https://[-0-9a-z]*\.trycloudflare.com' tunnel.log
print("--------------------------------------------\n")

# 4. Agora liga o motor na mesma célula
# Ele vai ficar rodando aqui e você vai ver o log subir
print("🚀 Ligando o ComfyUI... Aguarde o 'Starting server'!")
%cd /content/ComfyUI
!python main.py --preview-method auto --lowvram
```
